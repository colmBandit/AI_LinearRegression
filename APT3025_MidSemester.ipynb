```python
#QUESTION 2 B
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

#dataset
data = {
    'tv': [230.1, 44.5, 17.2, 151.5, 180.8],
    'radio': [37.8, 39.3, 45.9, 41.3, 10.8],
    'newspaper': [69.2, 45.1, 69.3, 58.5, 58.4],
    'sales': [22.1, 10.4, 9.3, 18.5, 12.9]
}

#the dataframe
df = pd.DataFrame(data)
print(df)
```

```python
#defining features and target
X = df[['tv', 'radio', 'newspaper']]
y = df['sales']

# splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# initializing linear regression model
reg = LinearRegression()
# training the model
reg.fit(X_train, y_train)
```

```python
#calculating the accuracy
accuracy = reg.score(X_test, y_test)
print(f"model accuracy: {accuracy:.2f}")
```

```python
# Part b Predicting sales for new data
new_data = pd.DataFrame([[165.5, 20.0, 35.4]], columns=X.columns)
predicted_sales = reg.predict(new_data)
print(f"Predicted Sales for TV, Radio, Newspaper: {predicted_sales[0]:.2f}")
```

```python
#part a
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

#loading the data
wine_data = pd.read_csv('/content/wine.csv')
```

```python
#part b
#selecting features and target variable
features = wine_data[['density', 'sulfates', 'residual_sugar']]
target = wine_data['high_quality']

#splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=0)
```

```python
#part c
#building the KNN model
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

#determine the best number of neighbors
train_accuracy = []
test_accuracy = []
```

```python
#part d
# Looping for different values of Knearest neigbor(1 to 20)
for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)

    train_accuracy.append(knn.score(X_train, y_train))
    test_accuracy.append(knn.score(X_test, y_test))
```

```python
#plotting
plt.figure(figsize=(10, 6))
plt.plot(range(1, 21), train_accuracy, label='training accuracy')
plt.plot(range(1, 21), test_accuracy, label='testing accuracy')
plt.title('KNN accuracy for different values of K')
plt.xlabel('number of neighbors')
plt.ylabel('accuracy')
plt.xticks(range(1, 21))
plt.legend()
plt.show()
```
